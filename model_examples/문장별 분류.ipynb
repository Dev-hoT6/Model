```python
import urllib.request
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
text['review'] = text.apply(lambda row: nltk.word_tokenize(row['review']), axis=1)
```

```python
print(text.head(5))
```

```python
text['review'] = text['review'].apply(lambda x: [WordNetLemmatizer().lemmatize(word, pos='v') for word in x])
print(text.head(5))
```

```python
tokenized_doc = text['review'].apply(lambda x: [word for word in x if len(word) > 1])
print(tokenized_doc[:5])
```

```python
# 역토큰화 (토큰화 작업을 되돌림)
detokenized_doc = []
for i in range(len(text)):
    t = ' '.join(tokenized_doc[i])
    detokenized_doc.append(t)

# 다시 text['headline_text']에 재저장
text['review'] = detokenized_doc
```

```python
text['review'][:10]
```

```python
# 상위 1,000개의 단어를 보존 
vectorizer = TfidfVectorizer(max_features= 1000)
X = vectorizer.fit_transform(text['review'])

# TF-IDF 행렬의 크기 확인
print('TF-IDF 행렬의 크기 :',X.shape)
```

```python
lda_model = LatentDirichletAllocation(n_components=10,learning_method='online',random_state=777,max_iter=1)
```

```python
lda_top = lda_model.fit_transform(X)
```

```python
print(lda_model.components_)
print(lda_model.components_.shape) 
```

```python
terms = vectorizer.get_feature_names_out()

def get_topics(components, feature_names, n=5):
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx+1), [(feature_names[i], topic[i].round(2)) for i in topic.argsort()[:-n - 1:-1]])

get_topics(lda_model.components_,terms)
```
